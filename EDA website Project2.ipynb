```python
'''Website Performance Analysis:
As of 2025 ,the total number of websites on the internet
is estimate to be around 110 to 120 crore

Over 60% of the web traffic now comes from mobile devices

Questions That we will be solving during this website data analysis
1) What patterns or trends can you observe in website sessions and user over time ?
2) Which marketing channel brought the highest number of user to the website,
and how can we use this insight to improve traffic from other sources?
3) which channel has the highest average enagemnet time,and what does that tell us about user behavior and content effectiveness?
4)How does engagemnet rate vary acroos differnet traffic channels?
5)Which channels are driving more engaged sessions compared to non-engaged ones,and what struggles can improve engagemnet is underperforming channels?
6) At what hours of the day does each channel drive the most traffic?

7) Is there any correlation between high traffic (sessions) and high engagement rate over time?'''
```

```python
#import python libraries

import numpy as np #numpy (slove complex data)
import pandas as pd #pandas(to give explorator form)
import matplotlib.pyplot as plt#matplot for gragh(visualizing)
%matplotlib inline
import seaborn as sns # seaborn
```

```python
#import csv file---csv=coma separated Values
df=pd.read_csv('/content/data-export (1) (1).csv',encoding='unicode_escape')
```

```python
df.shape #row X columns
```

```python
df.head()
```

```python
df.columns=df.iloc[0]#integer based location indexing #loc=label based indexing
df=df.drop(index=0).reset_index(drop=True)
df.columns=["channel group","DateHour","Users","Sessions","Engaged Session","Average engagement time per session","Engaged sessions per user","	Events per session	","Engagement rate","	Event count"]
```

```python
df.describe()
```

```python
df.head()
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"],format="%Y%m%d%H",errors='coerce')
```

```python
numeric_cols = df.columns.drop(["channel group", "DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
df["Hour"] = df["DateHour"].dt.hour

```

```python
df.head()
```

```python
df.describe()
```

```python
#sessions and user over time
sns.set(style="whitegrid")
```

```python
#1

plt.figure(figsize=(10,5))
df.groupby("DateHour") [["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and users over time")
plt.xlabel("DateHour")
plt.ylabel("count")
plt.show()
```

```python
#2
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="channel group", y="Users", estimator=np.sum, palette="viridis")
plt.title(" Total Users by Channel")
plt.xticks(rotation=45)
plt.show()

```

```python
#3 avreage enagment time by channel

plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="channel group", y="Average engagement time per session", estimator=np.mean, palette="viridis")
plt.title(" Average engagement time per session ")
plt.xlabel=("channel group")
plt.ylabel=("Average engagement time ")
plt.xticks (rotation=45)
plt.show()
```

```python
#4
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="channel group", y="Engagement rate",  palette="coolwarm")
plt.title(" Enageged Rate Distibution by channel ")
plt.xticks (rotation=45)
plt.show()
```

```python
#5
session_df = df.groupby("channel group") [["Sessions", "Engaged Session"]].sum().reset_index()

session_df["Non-Engaged"] = session_df["Sessions"] -session_df["Engaged Session"]

session_df_melted = session_df.melt(id_vars="channel group", value_vars=["Engaged Session" ,"Non-Engaged"])

plt.figure(figsize=(8, 5))

sns.barplot(data=session_df_melted, x="channel group", y="value", hue="variable")

plt.title(" Engaged vs Non-Engaged Sessions")

plt.xticks(rotation=45)

plt.show()
```

```python
#6
# traffic by hour and channel

heatmap_data = df.groupby(["Hour", "channel group"]) ["Sessions"].sum().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidths=.5, annot=True, fmt='.0f')
plt.title("Traffic by Hour and Channel")
plt.show()
```

```python
#7
df_plot = df.groupby("DateHour") [["Engagement rate", "Sessions"]].mean().reset_index()
plt.figure(figsize=(10, 5))
plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label="Engagement rate", color="green")
plt.plot(df_plot["DateHour"], df_plot["Sessions"], label="Sessions", color="blue")
plt.title("Engagement Rate vs Sessions Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
